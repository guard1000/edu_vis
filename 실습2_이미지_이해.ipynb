#### img를 gray scale로 바꿔보기

```python
img = cv2.imread(img_path) 
img_rgb = cv2.cvtColor(img,""" Your Answer""")    # 읽어온 이미지를 rgb로 바꿔봅니다.
img_gray = cv2.cvtColor(img, """ Your Answer""")  # 읽어온 이미지를 grayscale로 바꿔봅니다.

plt.figure(figsize=(20,12)) # plt 사이즈
plt.subplot(1, 3, 1) # 1행 2열에서 1번째 열
plt.imshow(img) # 컬러 - bgr

plt.subplot(1, 3, 2) # 1행 2열에서 2번째 열
plt.imshow(img_rgb, cmap='gray') # 컬러 - rgb

plt.subplot(1, 3, 3) # 1행 2열에서 3번째 열
plt.imshow(img_gray, cmap='gray') # 흑백

plt.show()
```
